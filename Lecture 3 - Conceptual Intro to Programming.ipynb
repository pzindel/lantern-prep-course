<h1><a id="c0">Preface</a></h1>
<div style="padding: 5% 5% 5% 5%">
    <p>This lecture is a beginner's guide to programming focusing on the conceptual aspect of topics. These topics will be practically implemented and covered using Python in the Python Programming course. This lecture is ideal for students to get early exposure and rudimentary understanding of the basics.</p>
</div>

<h1><a id="c1">Chapter 1: Introduction</a></h1>
    <div style="padding: 5% 5% 5% 5%">
    <p>What is programming and what is a program? In basic terms, a program is a set of instructions intend for a computer that tells the computer what it needs to do in order to solve some problem or complete some task. Programs can range from incredibly simple to extremely complex, and as a result, long programs are often broken down into smaller steps that tackle more individual and managable problems. By extension, programming is the act of writing the instructions that make up the program, and a programming language is the set of rules that dictates how the instructions should be written and translated for the computer to understand them.</p>
    <img src="images/3.1_baking.png" width=450 />
    <p>For example, suppose that you want your friend to bake a cake but, as it turns out, your friend has never set foot in a kitchen before and has no clue how to bake anything. To help your friend complete their task of baking a cake, you (the programmer) would have to write a clear set of instructions (the program) outlining in exact details the steps that your friend (the computer) needs to complete. The language that you are allowed to use for each instruction will be laid out in the programming language rulebook.</p>
    <p>Computers are not great at interpreting or making decisions for themselves which is why they require precise programming to tell them what to do, and while writing the necessary instructions for them to function may seem like a tedious task, there are many benefits to doing so. Unlike humans, computers have the ability to handle complex calcuations at breakneck speeds and repeat tasks again and again without getting tired or bored. Furthermore, once you write a program to solve a problem you can reuse the program without having to rewrite the instructions.</p>
    <img src="images/3.1_languages.png" width=150% />
    <figcaption style="text-align:center">A summary of the most commonly used programming languages.</figcaption>
    <p>There are a number of available programming languages that we can use to write our programs, each with their own pros and cons for when they should be used or not. Each language is free to set its own rules for allowable instructions (as long as they can be translated for the computer) and because of this the features and capabilities of languages will vary. The concepts discussed in the rest of this lecture are generally applicable to all programming languages, however preference is given to Python since that is the primary language used in subsequent Lantern courses.</p>
    </div>
</div>

<h1><a id="c2">Chapter 2: Syntax & Variables</a></h1>
<div style="padding: 5% 5% 5% 5%">
    <p>When writing programs, each programming language has its own set of rules that define how the code can be structured and which keywords are used for specific functionality. These rules are called the <b>syntax</b> of the language. When learning any new programming language, the new syntax is what you need to learn to be able to apply old principles in the new language.</p>
    <p>Programs are typically quite intricate and require completing numerous smaller tasks in order to achieve the main task of the whole program (similar to how you need to chop and mix ingredients as part of a recipe for a whole dish). To make accomplishing these tasks more managable, the synatx allows programs to be split into blocks of code that are chained together to form the whole program. In our baking analogy, this is similar to the recipe instruction being broken down into steps that you complete one at a time.</p>
    <p>The most commonly used syntax for separating code blocks is using curly brackets (braces) <code style="background-color:Moccasin">{}</code>. All code placed inside is the braces form a single block, and placing braces inside of braces forms nested code blocks. This brace syntax style is used in popular languages such as C, C#, and Java.</p>
    <p>In Python, whitespace indentation is used instead to divide code blocks and enforce the program's structure.</p><br />
    <img src="images/3.2_python_syntax.svg" width=550 />
    <p>The above image shows example code that defines two functions in Python. The colors show the different syntax elements of the language:</p>
    <ul>
        <li>keywords & operators are in blue;</li>
        <li>variables in black;</li>
        <li>numbers in green;</li>
        <li>strings in cyan.</li>
    </ul>
    <p>The example also shows the use of code blocks for different sections where each indentation shows a new code block. The functions are defined in the same code block and each has their own nested block. The code block in the second function is further broken down into more blocks:</p>
    <ul>
        <li>lines 5 to 20 forms the primary block inside of the <em>dotwrite</em> function;</li>
        <li>lines 9 to 12 forms a secondary (nested) block and lines 14 to 20 also forms a (separate) secondary block;</li>
        <li>each of the secondary blocks is contains two more nested blocks.</li>
    </ul>
    <p>The use of separate blocks of codes is sometimes a requirement, such as during loops and conditionals (see chapter 6), and other times the decision of using blocks is at the descretion of the programmer, like when using user-defined functions. The proper use of separated code blocks can make a problem easier to solve, the program easier to understand, and will allow you to reuse parts of a program in future problems and programs.</p>
    <p>As a benefit (or consequence) of the program being divided into code blocks of varying levels, any user-defined variables and functions have a <b>scope</b> that determines where these variables are defined and usable within the code.</p>
    <p>In our Python example above, line 15 defines a variable called <em>children</em> inside of a secondary code block of the <em>dotwrite</em> function. Due to scope, that variable is only available to use in the code block that it is defined in and any further nested code blocks (e.g. we wouldn't be able to use the variable in the secondary block on lines 9 to 12).</p>
    <p>Apart from the program structure, the code that you write will use a combination of keywords, operators, variables, and values throughout to form complete instructions.</p>
    <ul>
        <li>The language-dependent keywords are used to tell the computer what it is you would like to do. For example, if we want to create a user-defined function or initiate a loop block then we can use specific keywords to do so.</li>
        <li>The variables are user-named (the allowed naming rules are determined by the language) and used to keep track of data, data structures, and user-defined functions and objects. The variables allow the programmer to easily reference elements throughout the code.</li>
        <li>Values are the actual data values that a programmer can assign to variables and data structures. Languages have a set of natively available data types that the values can be (see chapter 3).</li>
        <li>Lastly, operators are similar to keywords but more universal between languages and focusing solely on manipulating data. There are operators that allow us to assign values to variables, operators to compare different values, and operators to perform arithmetic with our values (see chapter 5).</li>
    </ul>
    <p>Lastly, even though programming languages have strict rules there is still ambiguity in terms of how a programmer might name their variables and functions and just overall style their code. It's recommended that you look up a styling guide for language you intend to learn so that you can see what is the common practice among developers.</p>
    <p>The styling guide for Python can be found on the <a href="https://www.python.org/dev/peps/pep-0008/">official Python site</a>. Additionally, Google has set out their own <a href="https://google.github.io/styleguide/pyguide.html">Python styling guide</a> which has some minor variations.</p>
</div>

<h1><a id="c3">Chapter 3: Data Types</a></h1>
<div style="padding: 5% 5% 5% 5%">
    <p>For all programming languages, the type for any data or values in a program are either numerical or non-numerical. Exactly how those types are sub-divided and classified is specific to the language being used, but the following is generally true:</p>
    <ul>
        <li>Non-numerical data type refers to all character such as letters, special characters, and numbers (as characters). In general, single non-numerical characters are stored in a character (<code style="background-color:Moccasin">char</code>) type and multi-character data, typically called strings, is stored in a string (<code style="background-color:Moccasin">str</code>) type.</li>
        <ul>
            <li>Some languages, such as Java and C#, have implementations for both char and string, while others only have one implementation.</li>
            <li>C and C++ only have char data types for non-numerical data and utilize arrays of char types for storing multi-character strings.</li>
            <li>Python uses only multi-character string types for both.</li>
        </ul>
        <li>Numerical types includes all numbers that can be used directly in calculations (numbers as characters are excluded). These can be divided into different types depending on the format of the number and the range of possible values. In general there are three types: integers (whole numbers), floats (fractional/real numbers), bools (true or false values which can be stored as 1 or 0, respectively).</li>
        <ul>
            <li>Integers, or <code style="background-color:Moccasin">int</code>, types are usually 4 bytes of data giving a range of possible values from -2,147,483,648 to 2,148,483,647 ($2^{8*4}$ possible values). Some languages also have implementations for small integers (2 bytes) using <code style="background-color:Moccasin">short</code> types ($2^{8*2}$ possible values) and large integers (4 bytes) using <code style="background-color:Moccasin">long</code> types ($2^{8*8}$ possible values).</li>
            <li>Floating-point, or <code style="background-color:Moccasin">float</code>, types are usually 4 bytes of data where 1 bit is used for the sign, 8 bits are used for the exponent, and the remaining 27 bits are used for the significand. Some languages have implementations for double precision floating points, or <code style="background-color:Moccasin">double</code> types, that allow for 11 bits of exponent and 60 bits of significand.</li>
            <li>Boolean, or <code style="background-color:Moccasin">bool</code>, types typically use a single bit or byte to store a 0 or 1 that can be used as a state of false or true, respectively.</li>
        </ul>
    </ul>
    <img src="images/3.3_data_types.jfif" width=550 />
    <p>Note that in statically typed languages, such as Java and C, the data type must be identified when variables are declared (e.g. <code style="background-color:Moccasin">int my_int = 5;</code>) and in dynamically typed languages, such as Python, the compiling program interprets that data type automatically based on the values you pass to your variables (e.g. <code style="background-color:Moccasin">my_int = 5</code>).</p>
    <p>The reason why data types are important is to ensure that the operations that a programmer wishes to perform are valid for a given data type. For example, I cannot intuitively divide letters by a number and attempting to do so would result in an error. Note: by default that kind of mixed data type operation may not be defined however it would be possible to write create objects (see chapter 8) that are capable of more unusal operations.</p>
</div>

<h1><a id="c4">Chapter 4: Data Structures</a></h1>
<div style="padding:5% 5% 5% 5%">
    <img src="images/3.4_data_structures.png" width=750 />
    <p>If we want our program to work with lots of data or offer lots of customizable functionality than it will require storing lots of values. This can be a messy affair if we needed to store each value in a separate variable with a unique name. Luckily there are a number of commonly used data structures that can store collections of data in a way that we only need to track a single variable for the whole collection of data.</p>
    <p>The natively available data structures and their implementation are heavily language-dependent. For this section I place focus on those found in Python (since that that primary language used in Lantern courses) which are: lists, tuples, sets, and dictionaries.</p>
    <ul>
        <li>Lists are array data structures that can store ordered collections of data. In the case of Python, lists can store mix data types and are mutable, which means values can be reassigned in the structure.</li>
        <img src="images/3.4_lists.webp" width=450 />
        <li>Tuples are similar to list structures but they are immutable (cannot be modified once assigned).</li>
        <li>Sets are unordered data structures that only store one copy of duplicate values. This structure is primarily used when we are more interested in the existence of an object in the collection rather than the order or number of occurances. Sets provide useful functionality for finding unions and intersections.</li>
        <li>Dictionaries, or maps, are unordered data structures that act as a sort of address-book. With dictionaries we associate values with unique keys that we can use to retrieve the values. These key-value pairs make it easy for us to quickly retrieve and update the values as necessary.</li>
        <img src="images/3.4_dictionaries.png" width=450 />
    </ul>
    <p>Beyond the standard built-in data structure, users are also free to create their own custom data structure implementations. The extended set of data structures includes: linked lists, stacks, queues, trees, and graphs.</p>
    <ul>
        <li>Linked lists are oredered linear data structures similar to regular links, however in a linked list each item in the collection is composed of two parts: the data and a pointer to the next item in the list. The benefit of this design has to do with the way that normal lists are typically implemented. In a regular list, if we wish to add an item to the list we have to shift all of the items in the list to make room, whereas in a linked list we can just reassign the pointer.</li>
        <img src="images/3.4_linked_list.png" width=550 />
        <li>Queues and stacks are also similar to lists, however the order in which items can be placed and removed from queues and stacks is different from lists. In a queue, any item being added is added (enqueued) at the end of the queue and any item being removed (dequeued) is removed from the front (similar to a queue of people). Queue follow a first-in-first-out (FIFO) ordering. In a stack, any item being added is added at the top of the stack and any item being removed is also remove from the top (similar to a stack of papers). Stacks follow a last-in-first-out (LIFO) ordering.</li>
        <img src="images/3.4_queues_stacks.png" width=750 />
        <li>Graphs and trees are non-linear data structures. These structure consist of nodes and edge which are similar to the data and pointer elements for linked lists but are not restricted to a single on-way pointer. In a graph, any given node can have multiple edges connecting it to several other nodes and there is designated starting point (or root node). A tree is just a special case of a graph with a few more restrictions. In a tree there is a root node which has child nodes (which may also have children nodes etc.) and any given child node can only be connected to its direct parent node or direct child nodes.</li>
        <img src="images/3.4_graphs_trees.jpg" width=550 />
    </ul>
    <p>All of these different data structures has their own set of algorithms or operations which can be performed on them which are worth studying up on. The unique design of each structure means they all have their pros and cons which make the ideal for certain situations but not others. It's highly recommended to read-up on data structures and what set each apart from the next.</p>
</div>

<h1><a id="c6">Chapter 6: Conditionals & Loops</a></h1>
<div style="padding:5% 5% 5% 5%">
    <p>So far we've seen all of the basics of syntax and data handling, however we're still missing sveral key ingredients for programming to be practical. Two of them are loops and conditionals. Without either of these functionalities our applications would be quite bland and limited in what they could do as there would be no way to create variability with branching decisions and no way to easily repeat similar tasks.</p>
    <img src="images/3.6_conditional.svg" width=350 />
    <p>To "make decisions" within programs we can use conditionals called <b>if statements</b> that execute code blocks based on the conditions of variables. The leading statement checks if some condition is true, in which case the enclosed code block is executed, or false, in which case the code block is skipped and the program continues. If statements can also be chained together with other conditionals, else if statements, and a default running condition, else statement. A chained if-else code block can check several different conditions in an ordered manner and branch off into the desired nested block before returning to the program as a whole.</p>
    <p>For example, if we created an application with a user interface that allowed users to select a series of buttons and produce some output, then we could utilize if-else statements to produce a unique output that is dependent on the buttons that the user has pressed. For each button we could check its conditions (i.e. was it pushed or not) and given some combination of button pushes branch into a specific block of code.</p>
    <p>If statements typically utilized comparison operators between variables and booleans checks to set up the condtions for the branches.</p>
    <img src="images/3.6_while_loops.svg" width=350 />
    <p>If we want to repeat tasks similar tasks within our code, we can enclose the repeating code block into a loop. These looping blocks generally come in two flavours: <b>for loops</b>, and <b>while loops</b>.</p>
    <ul>
        <li>While loops are a type of looping code block with no definite number of iterations. A while loop has a fixed running condition, based on an if statement, that is checked on every iteration and if the condition is no longer met then the loop is exited (i.e. while loops are like repeating if blocks). While loops are particularly useful when the task at hand needs to be repeated an undetermined number of times.</li>
        <li>For loops are a type of looping code block that allows iterating over a sequence of data or a fixed number of times before exiting the loop block. This type of loops is useful when we wish to repeat a task a known number of times and with each iteration we can update variables (which allows for variability on every iteration).</li>
    </ul>
    <img src="images/3.6_for_loop.png" width=250 />
    <p>Both the conditional blocks and the loop blocks can be nested with each other any given number of times. This allows us to increase the complexity of a program as much as we require.</p>
</div>

<h1><a id="c7">Chapter 7: Functions</a></h1>
<div style="padding:5% 5% 5% 5%">
    <p>All of the elements that we've covered so far are essentail to write working and useful applications, however there is still one crucial missing component that grants us the ability to write highly resuable code.</p>
    <img src="images/3.7_functions.png" width=650 />
    <p>Functions are that component. With functions we can systemically break down our programs into a series of subroutines that perfom very specific tasks. From a high-level overview, these function act like miniature programs by taking some input data (see $x$ in the image above), performing some operation, and returning a new output value for us to use (see $f(x)$ above).</p>
    <p>By breaking down our program into smaller tasks and using functions to solve each task we gain a lot of flexibility that we otherwise wouldn't have. We previously mentioned that loop blocks can be used to repeat tasks however what we didn't mention was that loops are executed all at once. If I need to repeat the code inside of a loop again at a later time in my program I'm left with no choice but to write a new loop block (or really any block of code for that matter). With the addition of functions I can instead place similar code into a separate and well-defined block of code that has a variable name associated with it (as all regular function require). Then whenever I need to repeat that code block I can just call the function by its name.</p>
    <p>This kind of reusability is not only ideal within my program but also across any future programs I choose to write. As will be mentioned in chapter 10, programming languages allow us to import functions and classes from other files. This means we can easily repurpose code that we write now to save us time in the future. This also raises the importance of practicing good programming standards and design to ensure that your code can be reused will minimal modifications.</p>
    <p>Functions can be defined in such a way that also allows us to pass in variables to modify how the function block runs and what calculation it performs on which data. These functions can then either modify data structures directly, or return new data that can be assigned and utilized from the parent code block that called it.</p>
    <p>Furthermore, since functions can be designed to tackle problems of varying scale we can also design functions that nest within each other or get called from other functions.</p>
    <img src="images/3.7_functions_2.png" width=550 />
</div>

<h1><a id="c8">Chapter 8: Classes & Objects</a></h1>
<div style="padding:5% 5% 5% 5%">
    <p>When studying programming you will likely come by comparisons of different languages and hear about object-oriented programming (OOP). But what is OOP?</p>
    <p>Programming languages have two common ways of organizing data and executing instructions on that data. When a program is designed around functions that are separate from, and perform calculations on, data then it's called <b>procedure-oriented</b>. The alternate way of organizing your program is to combine data and functionality into a single variable called an object. This is called <b>object-oriented</b> programming, or OOP.</p>
    <p>A number of programming languages are only capable of procedural or object-oriented programming (hence why they are often compared for their use cases), and some languages are capable of both. C++ and Python are both examples of multi-paradigm lanuages that can do both.</p>
    <img src="images/3.8_oop.png" width=450 />
    <p>Classes and objects are two main components of object-oriented programming. A <b>class</b> defines the new data structure and an <b>object</b> is an instance of that class. For example, I could define an class called mug that defines the attributes of a mug (e.g. height, circumfrence, weight, capacity, material, etc.) and functions (typically called methods) of the mug (e.g. fill with liquid, drink, clean, etc.). Then suppose I have a polka-dotted mug that belongs to the mug class, the polka-dotted mug is an object which is an instance of the mug class that specifiy attribute values and shares the functionality of the class.</p>
    <p>Note the use of attribute and method to refer to a class/object characteristics and functionality, respectively, is intentional to help differentiate it from procedural programming.</p>
    <p>There are many great benefits of OOP including its efficiency and compact design by keep related data and functions together. The practices that really make OOP shine are <b>inheritance</b> and <b>polymorphism</b>. These two practices help provide heirarchy to different objects and provide further reusability to code based on that heirarchy.</p>
    <img src="images/3.8_inheritance.svg" width=350 />
    <p>Ihneritance is the mechanism where one class is based on another class with similar properties and functionality. When defining our subclass we can do so in a way that it will inherit its base attributes and methods from a superclass, and then we can add or overwrite as necessary to complete the implementation of our new class. For example, in the image above there is an already defined superclass of <code style="background-color:Moccasin">Animals</code> with general methods, such as move and eat, which all animals are capable of. If I wanted to create a class for <code style="background-color:Moccasin">Cat</code>, rather than starting completely from scratch, I could allow my class to inherit properties of the <code style="background-color:Moccasin">Animal</code> class and then define new methods, like <code style="background-color:Moccasin">meow()</code>, that only applies to cats.</p>
    <p>Polymorphism is a word which means "having many forms" and in programming this refers to the practice of functions and methods having multiple uses for different situations. In particular, when we are talking about polymorphism with respect to classes and objects, then its the practice of overriding methods inherited from a parent class to make it more applicatible to the subclass. For example, our <code style="background-color:Moccasin">Dog</code> class needs to override the <code style="background-color:Moccasin">move()</code> method of the parent class in a way that makes its movement specific to dogs. Similarly the <code style="background-color:Moccasin">Cat</code> class could override the <code style="background-color:Moccasin">move()</code> method in a way that is specific to cats. This mean the <code style="background-color:Moccasin">move()</code> method is polymorphic.</p>
    <p>Classes and objects are not something you have to worry too much about when you first begin to program, particularly in the data science field. However, using OOP is good practice so you should make sure to include it in your studies.</p>
</div>

<h1><a id="c9">Chapter 9: Errors & Exceptions</a></h1>
<div style="padding:5% 5% 5% 5%">
    <p>What happens if your program has some invalid statement or mistake? Regardless of how careful you are, you will likely encounter an error in your programming. As you practice programming you will become better at avoiding errors and much better at finding and correcting them.</p>
    <img src="images/3.9_errors.jpg" width=650 />
    <p>There are three kinds of errors: syntax errors, runtime errors, and logic errors.</p>
    <ul>
        <li>Syntax errors occur when the compiler finds a mistake in your program and it can't compile or execute it. This can happen when you use incorrect punctuation or refer to variables that have not yet been declared. Syntax errors are the easiest to find and fix since the compiler tells you where the mistake happened and what was wrong with the code.</li>
        <li>Runtime errors occur once the program has been compiled and a mistake is encountered while the program is running. This can happen when variables are storing the wrong data type or when a function is expecting data that is improperly passed to it. Runtime errors are a bit more difficult to handle than syntax errors but when the errors are encountered the program will output some error messege that can be used to fix the code.</li>
        <li>Logic errors, or bugs, are mistakes that occur where your program compiles and execute but does the wrong task. For example, if we wanted to add two numbers together but accidently subtracted them in the code, then since the computer doesn't know that there's a mistake, the code will compile and run without any errors but the output produced will be wrong. Logic errors are the most difficult to find and fix, and sometimes logic errors may go unnoticed since the program appears to function properly. Good ways to track down logic errors are:</li>
        <ul>
            <li>Think about what the program should be doing and what it may have done to produce the wrong results. This may help narrow down which part of the program is malfunctioning.</li>
            <li>Using print statements at key points in the program to confirm that the variables are set accordingly.</li>
            <li>Use a IDE with a dedicated debugging tool to step through your program and observe how variables are updated to find exactly where the mistake is.</li>
        </ul>
    </ul>
    <p>Exceptions are a method of defensive programming that allow programmers to create custom errors within their code when certain conditions are not met properly. For example, suppose you were writing a program to take a file as input and open the file for calculation. If the provided file path is invalid then your program will be incapable of producing the desired results. As a defense to this, the programmer can include an exception statement that throws an error in the event where the file path is invalid which will provide the user useful feedback about why the program failed its task. Defensive programming is always good to incorporate and is especially crucial to use whenever handling user input.</p>
</div>

<h1><a id="c10">Chapter 10: Libraries & APIs</a></h1>
<div style="padding:5% 5% 5% 5%">
    <p>The base programming languages only contain rudimentary functionality since a programming language's main purpose is to set out the rules for creating programs and interfacing with computer. As a result if programmer wish to complete a complex task they will have to implement a solutions from scratch. For example if you wanted to create a program that could perform matrix multiplication, then you would have to develop your own matrix data structure and methods for multiplying.</p>
    <p>This becomes a big problem when users want to complete their necessary tasks quickly without having to start from scratch each time. As a solution to this, programming languages are designed to allow users to import other program files and utilize the code that has already been written. These imported programs are called <b>libraries</b> or modules.</p>
    <img src="images/3.10_libraries.svg" width=450 />
    <figcaption style="text-align:center">Illustration of an application which uses libvorbisfile to play an Ogg Vorbis file</figcaption>
    <p>Libraries can be written the developers of the language, developers in the community, or even you. The benefit of using libraries is that they expand the available functionality and allow programmers to select necessary and relevant code to quickly create large-scale applications. Furthermore, having extra functionality broken down across many different libraries bring the benefit of a light-weight program that only contains the absolutely needed code. In contrast if every function imaginable were included in the language by default then it would be quite bloated and slow.</p>
    <p>One of the major advantages of the Python language is its vast community and libraries available for use. This means programmers can quickly prototype applications by utilizing the functions and objects that have been defined by other developers without having to write their own implementation from scratch. The Python Programming course will go into more depth about the essential libraries to use when working with data.</p>
