        <figcaption style="text-align:center">Screenshot of a Bash terminal and common commands.</figcaption>

        <p>The full Bash reference manual with definitions, a list of built-in commands, and features can be found over on the <a href="https://www.gnu.org/software/bash/manual/bash.html">GNU site</a>.
        <div style="padding: 2% 2% 2% 2%; margin: 2% 0% 2% 0%; background-color: Moccasin;">

            <div style="padding: 0% 5% 0% 5%">
                <h5>Windows</h5>
                <ol>
                    <li>Download the Git for Windows <a href="https://git-for-windows.github.io/">installer</a>.</li>
                    <li>Run the installer and follow the installation steps in accordance with the following:</li>
                    <ol>
                        <li><b>Make sure to keep "Use Git from the Windows Command Prompt" selected!</b> If you forget to do this certain programs may not work properly. If this happens rerun the installer and select the appropriate option.</li>
                        <li><b>Make sure to keep "Checkout Windows-style, commit Unix-style line endings" selected!</b></li>
                        <li><b>Make sure to keep "Use Windows' default console window" selected!</b></li>
                    </ol>
                    <li>If your "HOME" environment vairable is not set (or you don't know what this is):</li>
                    <ol>
                        <li>Open command prompt (Open Start Menu then type <code style="color:DarkGreen">cmd</code> and press Enter).</li>
                        <li>Type the following line into the command prompt window exactly as shown: <code style="color:DarkGreen">setx HOME "%USERPROFILE%"</code></li>
                        <li>Press Enter, you should see <code style="color:DarkGreen">SUCCESS: Specified value was saved.</code></li>
                        <li>Quit command prompt by typing <code style="color:DarkGreen">exit</code> then pressing Enter.</li>

    <h3><a id="s1.5">&sect;1.5 Setting the PATH Environment Variable</a></h3><!--####################################-->
    <div style="padding: 1% 5% 0% 5%; background-color: White;">
        <p>Once you've installed programs that you wish to use through command-line, it may be necessary to update the <b>PATH</b> environment variable with the installation directories of the new programs. Setting this environment variable appropriately tells Bash where certain programs are located which makes them easier to access and use.</p>
        <p>For example, if I've installed Notepad++ to use as my primary text editor I can try to open the program from the Bash command-line using <code style="color:DarkGreen">notepad++</code> or <code style="color:DarkGreen">notepad++.exe</code>. However, since the PATH variable is not set to include this program, I will receive the error "command not found". I could use the full directory path that the program was installed to, i.e. <code style="color:DarkGreen">/C/'Program Files (x86)'/Notepad++/notepad++.exe</code>, but this is inefficient and tedious. Instead I can add the installation folder to the PATH variable which will allow me to use the shortforms.</p>
        <p>There are several ways to add a folder to the PATH variable. The easiest method is to open a Bash terminal and open the <b><em>.bashrc</em></b> file using your currently installed editor, e.g. <code style="color:DarkGreen">nano .bashrc</code>. The editor should open the text file which may already contain some text or be completely empty. On a new line add <code style="color:DarkGreen">export PATH=&lt;dir&gt;:&dollar;PATH</code> where <em>&lt;dir&gt;</em> is replaced with the full directory path to the program.</p>
        <p>In our Notepad++ example I would add the line <code style="color:DarkGreen">export PATH=/C/'Program Files (x86)'/Notepad++:&dollar;PATH</code>.</p>
        <p>Once you've added the line you can save the file and restart Bash to start using the shortform names.</p>
        <p>Note that we can also set an <b>alias</b> for our program that will allow us to use an alternate name to open a program. To do this we can go open the <b><em>.bash_profile</em></b> file in our text editor, e.g. <code style="color:DarkGreen">nano .bash_profile</code> and add the line <code style="color:DarkGreen">alias &lt;alias name&gt;=&lt;original name&gt;</code>. For our Notepad++ example we could set an alias as <code style="color:DarkGreen">alias npp=notepad++.exe</code> to open the program with the <code style="color:DarkGreen">npp</code> shortform.</p>
    </div>

    <h3><a id="s1.7">&sect;1.7 Setting up Python Virtual Environments</a></h3><!--######################################-->
    <div style="padding: 1% 5% 0% 5%; background-color: White;">
        <p>Different Python applications will require different extra packages and modules. Sometimes an application will need a specific version of a library or certain packages my conflict. A solution to this problem is to create vitual environments with separate installed packages. This is possible <a href="https://docs.python.org/3/tutorial/venv.html">natively to Python</a> but if Anaconda is installed it is recommended to use its virtual environment implementation.</p>
        <p>Similar to package installation, setting up a new environment and adding packages to install can be done from a Bash terminal or using the Anaconda Nagivator program. To do so <a href="https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html">from the Bash terminal</a>:</p>
        <ol>
            <li>Update conda to the latest version using <code style="color:DarkGreen">conda update conda</code> and typing <code style="color:DarkGreen">y</code> and pressing Enter when prompted.</li>
            <li>Create a virtual environment using <code style="color:DarkGreen">conda create -n envname</code> where <em>envname</em> is the name you wish to assign your new virtual environment.</li>
            <li>Get a list of available environments using <code style="color:DarkGreen">conda info -e</code> command.</li>
            <li>Activate your desired virtual environment using <code style="color:DarkGreen">source activate envname</code>.</li>
            <li>Install the required Python packages to the new environment using <code style="color:DarkGreen">conda install -n envname package</code> where package is the package name as mentioned above.</li>
            <li>Once you're done with a virtual environment and no longer need it, you can delete the environment by deactivating it, <code style="color:DarkGreen">conda deactivate</code>, and removing the environment and its packages, <code style="color:DarkGreen">conda remove -n envname --all</code>.</li>
        </ol>
    </div>

<div style="padding: 5% 5% 5% 5%">
    <p>Now that we have all of the applications installed and set up we have 3 different ways that we can work with Python. The following sections will give an overview of the different working environments at our disposal and the situations where you might prefer to use one over the other.</p>
    <h3><a id="s2.1">&sect;2.1 Jupyter Notebook & JupyterLab</a></h3><!--###############################################-->
    <div style="padding: 1% 5% 0% 5%">
        <img src="images/1.2_jupyter_notebook_screenshot_dashboard.png" width=750 />
        <figcaption style="text-align:center">Screenshot of the Jupyter Notebook dashboard.</figcaption>
        <p>The primary distinction of the Anaconda package is its <a href="https://jupyter-notebook.readthedocs.io/en/stable/">Jupyter Notebook</a> environment. Jupyter Notebook is a web application that allows users to mix live code, images, equations, narrative text, and full HTML markdown into a single file (This lecture was written in Jupyter Notebook).</p>
        <p>To use Jupyter Notebook you need an up-to-date browser (as mentioned in section 1.4) to properly render the user interface and functionality. You can launch Jupyter Notebook from the Anaconda Navigator or through a Bash terminal using the command <code style="DarkGreen">jupyter notebook</code> and it will open the program dashboard in your default web browser.</p>
        <p>The Notebook dashboard will allow you to navigate your file directory and perform basic file management task such as duplicating, renaming, moving, and deleting. In this dashboard you will be able to create new notebook files and load present notebooks for viewing and editing. The program uses the .ipynb format for its notebook documents. Jupyter Notebook also allows users to view other files directly from the dashboard and edit basic text files.</p>
        <p>Once you've selected a notebook to edit, the notebook will open in the Jupyter Notebook Editor where you will be able to write your code and markdown (formattable text) inside of cells. <b>To learn more about the specific elements withing the Notebook Editor, you can go through the user interface tour by selecting <em>Help</em> in the menubar at the top and then selecting <em>User Interface Tour</em>.</b></p>
        <p>The notebooks are broken down into modular cells that allow you to write and run code one block at a time. There are two main types of cells: code and markdown. The code cells are where you will be writing you Python code and the markdown cells are where you will be able to place formatted narrative text and HTML (along with LaTeX formatted equations). When a cell is in edit mode it will be indicated by a green outline and a small pencil icon on the top right of the interface. When a code cell is executed it will display any printed output in the space below the cell.</p>
        <img src="images/1.2_jupyter_notebook_screenshot_edit.png" width=750 />
        <figcaption style="text-align:center">Screenshot of a notebook being edited.</figcaption>
        <p>Within a notebook, all of the program data, including variables and data, is shared. This means you can split up your code into multiple cells and still have access to any variables defined in previous cells (as long as they are executed in the correct order). Note that this is not true across different notebooks, i.e. any data and variables that we wish to use in the current notebook must also be defined in the current notebook or in an imported file/library.</p>
        <p>The benefits provided by the Jupyter Notebook environment comes from the ability for users to interlace their code cells with the markdown cells, and also run code in a modular fashion with output values being printed after each cell as it's executed. This feature is particularly useful for those wishing to perform data analytics in Python as users can get continuous feedback on their code and the current state of the program. It simplifies the data exploration and plotting processes. Additionally, users are able to make note of their observations and conclusions in markdown cells as they progress through their workflow.</p>
        <img src="images/1.2_jupyter_lab_screenshot_dashboard.png" width=750 />
        <figcaption style="text-align:center">Screenshot of the Jupyter Lab landing page shows the updated layout with the same functionality.</figcaption>
        <p><a href="https://jupyterlab.readthedocs.io/en/latest/">JupyterLab</a> is a next-gen interface for Jupyter notebooks. It extends and improves the functionality of Jupyter Notebook, condensing the open notebooks and interface into a singular tab with a flexible and arrangable user interface. JupyterLab can be lauched similar to Jupyter Notebook, either through the Anaconda Navigator or using the command <code style="DarkGreen">jupyter lab</code>.</p>
        <p>For a more comprehensive look at Jupyter Notebooks, review the brief documentation on:
            <ul>
                <li><a href="https://nbviewer.jupyter.org/github/jupyter/notebook/blob/master/docs/source/examples/Notebook/Notebook%20Basics.ipynb">Notebook Basics</a></li>
                <li><a href="https://nbviewer.jupyter.org/github/jupyter/notebook/blob/master/docs/source/examples/Notebook/Running%20Code.ipynb">Running Code</a></li>
                <li><a href="https://nbviewer.jupyter.org/github/jupyter/notebook/blob/master/docs/source/examples/Notebook/Working%20With%20Markdown%20Cells.ipynb">Markdown Cells</a></li>
                <li><a href="https://nbviewer.jupyter.org/github/jupyter/notebook/blob/master/docs/source/examples/Notebook/Typesetting%20Equations.ipynb">Typesetting Equations</a></li>
        </ul>
    </div>
    <h3><a id="s2.2">&sect;2.2 Spyder</a></h3><!--##################################################-->
    <div style="padding: 1% 5% 0% 5%">
        <img src="images/1.2_spyder_screenshot.png">
        <figcaption style="text-align:center">Screenshot of the Spyder IDE.</figcaption>
        <p>The second coding environment provided by Anaconda is the Spyder IDE. Unlike Jupyter, Spyder is a more traditional IDE that you would commonly use for other compiled programming languages. The application doesn't allow users to add well formatted markdown and equations inbetween their code, however the usual coding comment methods are still available for use.</p>
        <p>Spyder's user interface provides three main windows: the editor (left), the object inspector (top right), and the console (bottom right). Within the editor window users will be able to select different scripts and write their code inside of them in a single block. The current script can then be run using the hotbar at the top which will execute the chosen program in the console window and will display any console output along the way. Any objects and variables that are loaded into memory by the program can be explored using the top right window.</p>
        <p>The Spyder documentation contain a concise <a href="https://docs.spyder-ide.org/current/quickstart.html">quickstart tour of the IDE</a> for new users.</p>
        <p>Since Spyder is formatted like a traditional IDE its intended use is also for more traditional programming. Programs are written in files, containing only Python code, which can be run from any Python console or program. The code is written all in a single cell and any time you execute a script in Spyder, all lines of code in the file are executed (versus Jupyter Notebook's more modular execution).</p>
        <p>The programming style that Spyder caters to is that which is more typically used for software development rather than data analysis. Furthermore, Spyder benefits over Jupyter through its build-in debugger which allows you to step through your program, one line at a time, and figure out where any mistakes are occuring.</p>
        <p>The <a href="https://youtube.com/playlist?list=PLPonohdiDqg9epClEcXoAPUiK0pN5eRoc">Spyder IDE tutorial videos</a> are a good place to get started with familiarizing yourself with the environment and working in it.</p>
    </div>
    <h3><a id="s2.3">&sect;2.3 Text Editor + CLI</a></h3><!--##################################################-->
    <div style="padding: 1% 5% 0% 5%">
        <p>The third option we have for programming in Python is to use a text editor to write our code and the command-line interface (CLI) to execute our scripts. This method of programming is similar to using Spyder but without any of the extra tools. As a result, this method is less suited to writing long and intricate programs since debugging code and fixing mistakes can be a more difficult process.</p>
        <p>Using Python through the CLI bring us two main benefits. If we wish to write a short Python script to prototype or complete a simple task than using the CLI can be a quick and easy method for doing so. The second, and main, benefit of using the CLI is when it comes time to run Python scripts. When properly utilized, the CLI allows users to provide extra inputs that can alter the way Python programs run (assuming the program was written in a way to handle the given extra input). This is also useful when you wish to automate a process and have outputs from one program be used as inputs in another program, which is something is outside of the intended scope of Spyder.</p>
        <p>To start writing Python code, you can open up a Python file with your preferred text editor using the command <code style="color:DarKGreen">nano my_script.py</code> (in this case <em>nano</em> is the editor being used and <em>my_script.py</em> is the Python file I want to code in). This will open the file an allow you begin programming. Once you are finished and have saved and exited the text editor, you can run the new Python script from the command-line using the code <code style="color:DarkGreen">python my_script.py</code>. Note: in some cases the alias for calling Python is <code style="color:DarkGreen">python3</code>.</p>
        <p>While in the command-line, you can also just call <code style="color:DarkGreen">python</code> without a script to open an interactive Python interpreter that will allow you to run Python code one line at a time. Be aware that any code you write directly in the interpreter like this will be lost once you exit the interpreter.</p>
    </div>
